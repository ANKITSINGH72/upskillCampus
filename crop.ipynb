```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# Step 1: Load the data
new_df = pd.read_csv('cleaned_agriculture_data.csv')

# Step 2: Data Cleaning
# Ensure all numeric columns are actually numeric
for column in new_df.columns:
    if new_df[column].dtype == 'object':
        try:
            new_df[column] = new_df[column].astype(float)
        except ValueError:
            pass  # Ignore conversion errors, those columns will be treated as categorical

# Handle missing values
new_df.dropna(axis=1, thresh=int(0.5 * len(new_df)), inplace=True)  # Drop columns with more than 50% missing values
new_df.fillna(new_df.mean(numeric_only=True), inplace=True)  # Fill remaining numeric missing values with mean

# Step 3: Encoding categorical features
label_encoders = {}
for column in new_df.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    new_df[column] = label_encoders[column].fit_transform(new_df[column])

# Step 4: Feature Selection
features = new_df.drop(['Production 2010-11'], axis=1)  # Drop the target column from features
target = new_df['Production 2010-11']  # Target variable (adjust this based on your target year)

# Step 5: Split the data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Step 6: Model Building
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 7: Predictions and Evaluation
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f'Mean Squared Error: {mse}')
print(f'RÂ² Score: {r2}')
```

```python
import pandas as pd

# Assuming 'model' is your trained model and 'features' is your feature DataFrame
feature_importances = pd.Series(model.feature_importances_, index=features.columns).sort_values(ascending=False)
print("Feature Importances:")
print(feature_importances)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define the model
rf = RandomForestRegressor()

# Define hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(features, target)

# Best parameters and score
print("Best Parameters:")
print(grid_search.best_params_)
print("Best Score:")
print(grid_search.best_score_)
```

```python
from sklearn.model_selection import cross_val_score

# Evaluate model with cross-validation
cv_scores = cross_val_score(model, features, target, cv=5, scoring='r2')
print(f'Cross-Validation RÂ² Scores: {cv_scores}')
print(f'Mean CV RÂ² Score: {cv_scores.mean()}')
```
